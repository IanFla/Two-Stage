```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime as dt
from pandas_datareader import DataReader as DR
import seaborn as sb
import numdifftools as nd
from wquantiles import quantile

from scipy.stats import multivariate_normal as multi_norm
from scipy.stats import multivariate_t as multi_t
from scipy.stats import norm,t,truncnorm
from scipy.spatial import Delaunay as TRI
from scipy.interpolate import LinearNDInterpolator as ITP
from scipy.optimize import minimize

from sklearn.linear_model import LinearRegression as Linear
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neighbors import KernelDensity as KDE

import warnings
warnings.filterwarnings("ignore")
```

```python
data=DR('^GSPC','yahoo',dt(2010,9,29),dt(2011,7,14))
returns=pd.DataFrame(100*np.diff(np.log(data['Adj Close'])),columns=['dlr'])
returns.index=data.index.values[1:data.index.values.shape[0]]

returns['dlr'].plot()
plt.show()
returns=np.array(returns['dlr'])
```

```python
def posterior_prediction(pars,pre=False):
    good=(np.exp(pars[:,0])<=1e300)
    good&=(pars[:,1]>0)*(pars[:,2]>0)*(pars[:,1]+pars[:,2]<1)
    neglogpdfpT=0.5*((pars[:,0]-prior_pars[0])/prior_pars[1])**2
    neglogpdfpT[~good]=np.inf
    HTd=np.ones(pars.shape[0])
    HTd[good]=np.exp(pars[good,0])+pars[good,1]*y0**2+pars[good,2]*h0
    for i in range(T):
        neglogpdfpT+=0.5*(YT[i]**2/HTd+np.log(HTd))
        HTd[good]=np.exp(pars[good,0])+pars[good,1]*YT[i]**2+pars[good,2]*HTd[good]
        
    if pre:
        Yd=np.zeros([pars.shape[0],d])
        pdfpdpt=np.zeros(pars.shape[0])
        good=(np.exp(-neglogpdfpT)!=0)
        pdfpdpt[good]=1
        if d>1:
            for i in range(d-1):
                Yd[good,i]=norm.rvs(scale=np.sqrt(HTd[good]))
                pdfpdpt[good]*=norm.pdf(x=Yd[good,i],scale=np.sqrt(HTd[good]))
                HTd[good]=np.exp(pars[good,0])+pars[good,1]*Yd[good,i]**2+pars[good,2]*HTd[good]
            
        return neglogpdfpT,HTd,Yd,pdfpdpt
    else:
        return neglogpdfpT

def MLE():
    cons=({'type':'ineq',\
           'fun':lambda pars: np.array([pars[1],pars[2],1-pars[1]+pars[2]]),\
           'jac':lambda x: np.array([[0,1,0],[0,0,1],[0,-1,-1]])})
    target=lambda pars: posterior_prediction(pars.reshape([1,3]))
    res=minimize(target,[0,0.1,0.7],method='SLSQP',constraints=cons,\
                 options={'maxiter':1000,'ftol':1e-100,'disp':False})
    mu=res['x']
    Sigma=np.linalg.inv(nd.Hessian(target)(mu))
    Sigma[:,0]*=2
    Sigma[0,:]*=2
    return mu,Sigma

def estimation(RYd,W,p,ESS=False):
    VaR=quantile(RYd,W,p)
    w=W/np.sum(W)
    avar=np.sum((w*((RYd<=VaR)-p))**2)*RYd.size
    if ESS:
        print('ESS for',str(p)+':')
        Wf=W*(RYd<=VaR)
        wf=Wf/np.sum(Wf)
        ne=1/np.sum(w**2)
        nef=1/np.sum(wf**2)
        print('ne:',ne,'nef:',nef)
        W2=W**2
        w2=W2/np.sum(W2)
        Wf2=(W*((RYd<=VaR)-p))**2
        wf2=Wf2/np.sum(Wf2)
        ne2=1/np.sum(w2**2)
        nef2=1/np.sum(wf2**2)
        print('ne2:',ne2,'nef2:',nef2)
        
    return VaR,avar

def MLE_estimation(RYd,pdfp,pdfqs):
    pdfg=pdfqs[1:]-pdfqs[0]
    pdfq=pdfqs.mean(axis=0)
    target=lambda zeta: -np.sum(np.log(pdfq+zeta.reshape([1,-1]).dot(pdfg).flatten()))
    gradient=lambda zeta: -np.sum(pdfg/(pdfq+zeta.reshape([1,-1]).dot(pdfg).flatten()),axis=1)
    zeta0=np.zeros(pdfg.shape[0])
    cons=({'type':'ineq',\
           'fun':lambda zeta: pdfq+zeta.reshape([1,-1]).dot(pdfg).flatten(),\
           'jac':lambda zeta: pdfg.T})
    res=minimize(target,zeta0,method='SLSQP',jac=gradient,constraints=cons,\
                 options={'maxiter':1000,'ftol':1e-100,'disp':False})
    if res['success']:
        zeta=res['x']
        tmp=np.append(-zeta.sum(),zeta)
        print('MLE',res['nit'],'alpha:',np.ones(pdfqs.shape[0])/pdfqs.shape[0]+tmp)
        W=pdfp/(pdfq+zeta.reshape([1,-1]).dot(pdfg).flatten())
        print('MLE',RYd.size,'1~4:','VaR0.05:',estimation(RYd,W,0.05)[0],'VaR0.01:',estimation(RYd,W,0.01)[0])
    else:
        print('MLE fail')
    
def proposal(typ,mode,HTd,rvs=None,VaR=None,RYdpart=None):
    if typ=='rvs':
        if mode==1:
            rvs=norm.rvs(scale=np.sqrt(HTd))
        elif mode==2:
            rvs=norm.rvs(loc=-np.sqrt(HTd),scale=np.sqrt(HTd))
        elif mode==3:
            rvs=truncnorm.rvs(scale=np.sqrt(HTd),a=-np.inf,b=(VaR-RYdpart)/np.sqrt(HTd))
        elif mode==4:
            rvs=truncnorm.rvs(loc=-np.sqrt(HTd),scale=np.sqrt(HTd),a=-np.inf,b=(VaR-RYdpart)/np.sqrt(HTd)+1)
        else:
            print('mode error')
            
        return rvs
    elif typ=='pdf':
        if mode==1:
            pdf=norm.pdf(x=rvs,scale=np.sqrt(HTd))
        elif mode==2:
            pdf=norm.pdf(x=rvs,loc=-np.sqrt(HTd),scale=np.sqrt(HTd))
        elif mode==3:
            pdf=truncnorm.pdf(x=rvs,scale=np.sqrt(HTd),a=-np.inf,b=(VaR-RYdpart)/np.sqrt(HTd))
        elif mode==4:
            pdf=truncnorm.pdf(x=rvs,loc=-np.sqrt(HTd),scale=np.sqrt(HTd),a=-np.inf,b=(VaR-RYdpart)/np.sqrt(HTd)+1)
        else:
            print('mode error')
            
        return pdf
    else:
        print('type error')

def pilot():
    parsN=multi_norm.rvs(size=6*n0,mean=mu,cov=Sigma)
    parst=np.array([t.rvs(size=2*n0,df=1,loc=mu[i],scale=np.sqrt(Sigma[i,i])) for i in range(3)]).T
    pars=np.vstack([parsN[:2*n0],parst,parsN[2*n0:]])
    pdfN=multi_norm.pdf(x=pars,mean=mu,cov=Sigma)
    pdft=np.prod([t.pdf(x=pars[:,i],df=1,loc=mu[i],scale=np.sqrt(Sigma[i,i])) for i in range(3)],axis=0)
    
    neglogpdfpT,HTd,Yd=posterior_prediction(pars,True)[:3]
    pdfpT=np.exp(-neglogpdfpT)
    good=(neglogpdfpT!=np.inf)
    print('Samples in feasible region:',np.mean(good))
    
    # Step 1
    rvs1=proposal('rvs',1,HTd[:n0])
    rvs2=proposal('rvs',2,HTd[n0:2*n0])
    rvs3=proposal('rvs',1,HTd[2*n0:3*n0])
    rvs4=proposal('rvs',2,HTd[3*n0:4*n0])
    Yd[:4*n0,d-1]=np.hstack([rvs1,rvs2,rvs3,rvs4])
    RYd=Yd[:4*n0].sum(axis=1)
    
    pdfq1=proposal('pdf',1,HTd[:4*n0],Yd[:4*n0,d-1])
    pdfq2=proposal('pdf',2,HTd[:4*n0],Yd[:4*n0,d-1])
    pdfq1N=pdfN[:4*n0]*pdfq1
    pdfq2N=pdfN[:4*n0]*pdfq2
    pdfq1t=pdft[:4*n0]*pdfq1
    pdfq2t=pdft[:4*n0]*pdfq2
    
    pdfqs=np.array([pdfq1N,pdfq2N,pdfq1t,pdfq2t])
    pdfp=pdfpT[:4*n0]*pdfq1
    W=pdfp/pdfqs.mean(axis=0)
    VaR1=estimation(RYd,W,0.05)[0]
    print(4*n0,'1~4:','VaR0.05:',estimation(RYd,W,0.05),'VaR0.01:',estimation(RYd,W,0.01))
    
    # Step 2
    rvs5=proposal('rvs',3,HTd[4*n0:5*n0],VaR=0.8*VaR1,RYdpart=Yd[4*n0:5*n0,:d-1].sum(axis=1))
    rvs6=proposal('rvs',4,HTd[5*n0:6*n0],VaR=0.8*VaR1,RYdpart=Yd[5*n0:6*n0,:d-1].sum(axis=1))
    Yd[4*n0:6*n0,d-1]=np.hstack([rvs5,rvs6])
    RYd=Yd[:6*n0].sum(axis=1)
    
    pdfq1=proposal('pdf',1,HTd[:6*n0],Yd[:6*n0,d-1])
    pdfq2=proposal('pdf',2,HTd[:6*n0],Yd[:6*n0,d-1])
    pdfq1N=pdfN[:6*n0]*pdfq1
    pdfq2N=pdfN[:6*n0]*pdfq2
    pdfq1t=pdft[:6*n0]*pdfq1
    pdfq2t=pdft[:6*n0]*pdfq2
    pdfq5N=pdfN[:6*n0]*proposal('pdf',3,HTd[:6*n0],Yd[:6*n0,d-1],VaR=0.8*VaR1,RYdpart=Yd[:6*n0,:d-1].sum(axis=1))
    pdfq6N=pdfN[:6*n0]*proposal('pdf',4,HTd[:6*n0],Yd[:6*n0,d-1],VaR=0.8*VaR1,RYdpart=Yd[:6*n0,:d-1].sum(axis=1))
    
    pdfqs=np.array([pdfq1N,pdfq2N,pdfq1t,pdfq2t,pdfq5N,pdfq6N])
    pdfp=pdfpT[:6*n0]*pdfq1
    W=pdfp/pdfqs.mean(axis=0)
    VaR2=estimation(RYd,W,0.01)[0]
    print(6*n0,'1~6:','VaR0.05:',estimation(RYd,W,0.05),'VaR0.01:',estimation(RYd,W,0.01))
    
    # Step 3
    rvs7=proposal('rvs',3,HTd[6*n0:7*n0],VaR=0.8*VaR2,RYdpart=Yd[6*n0:7*n0,:d-1].sum(axis=1))
    rvs8=proposal('rvs',4,HTd[7*n0:8*n0],VaR=0.8*VaR2,RYdpart=Yd[7*n0:8*n0,:d-1].sum(axis=1))
    Yd[6*n0:8*n0,d-1]=np.hstack([rvs7,rvs8])
    RYd=Yd.sum(axis=1)
    
    pdfq1=proposal('pdf',1,HTd,Yd[:,d-1])
    pdfq2=proposal('pdf',2,HTd,Yd[:,d-1])
    pdfq1N=pdfN*pdfq1
    pdfq2N=pdfN*pdfq2
    pdfq1t=pdft*pdfq1
    pdfq2t=pdft*pdfq2
    pdfq5N=pdfN*proposal('pdf',3,HTd,Yd[:,d-1],VaR=0.8*VaR1,RYdpart=Yd[:,:d-1].sum(axis=1))
    pdfq6N=pdfN*proposal('pdf',4,HTd,Yd[:,d-1],VaR=0.8*VaR1,RYdpart=Yd[:,:d-1].sum(axis=1))
    pdfq7N=pdfN*proposal('pdf',3,HTd,Yd[:,d-1],VaR=0.8*VaR2,RYdpart=Yd[:,:d-1].sum(axis=1))
    pdfq8N=pdfN*proposal('pdf',4,HTd,Yd[:,d-1],VaR=0.8*VaR2,RYdpart=Yd[:,:d-1].sum(axis=1))
    
    pdfqs=np.array([pdfq1N,pdfq2N,pdfq1t,pdfq2t,pdfq5N,pdfq6N,pdfq7N,pdfq8N])
    pdfp=pdfpT*pdfq1
    W=pdfp/pdfqs.mean(axis=0)
    print(8*n0,'1~8:','VaR0.05:',estimation(RYd,W,0.05),'VaR0.01:',estimation(RYd,W,0.01))
    
    return pars,Yd,good
```

```python
h0=np.std(returns)
y0=returns[0]
YT=returns[1:]
T=YT.size
prior_pars=[-1,2]
truth=np.array([[-1.333,-1.895],[-1.886,-2.771],[-2.996,-4.424]])
mu,Sigma=MLE()
diag=np.diag(1/np.sqrt(np.diag(Sigma)))
print(diag.dot(Sigma).dot(diag))
```
